```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
Storedf = pd.read_csv('Mall_Customers.csv')
Storedf.head()
```

```python
Storedf.describe()
```

```python
X = Storedf.iloc[: ,[3,4]].values #iloc is used to extract the data based on index no
```

```python
from sklearn.cluster import KMeans
wcsos = [] #within cluster sum of squares
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i , init='k-means++', random_state=0)
    kmeans.fit(X)
    wcsos.append(kmeans.inertia_) # formula to segregate data into clusters
```

```python
plt.plot(range(1,11),wcsos)
plt.title("Elbow plot method")
plt.xlabel('No of clusters')
plt.ylabel("wcsos values")
plt.show()
```

```python
#Kmeans model
kmeansmodel = KMeans(n_clusters=5,init='k-means++',random_state=0)
```

```python

#Predict y
y_kmeans = kmeansmodel.fit_predict(X)
```

```python
#Scatter Plot
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1],s=80 , c ="red", label="Customer 1")
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1], s=80 , c="blue", label = "Customer 2")
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1], s=80 , c="green", label = "Customer 3")
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1], s=80 , c="yellow", label = "Customer 4")
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1], s=80 , c="pink", label = "Customer 5")

plt.title("Clusters of the Customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
```
